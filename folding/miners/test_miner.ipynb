```python
import os 
import time
import argparse 
import pandas as pd 

import bittensor as bt 

from folding.protocol import FoldingSynapse
from folding_miner import FoldingMiner

from folding.utils.config import add_miner_args, add_args
```

```python
parser = argparse.ArgumentParser()
# bt.wallet.add_args(parser)
# bt.subtensor.add_args(parser)
# bt.logging.add_args(parser)
# bt.axon.add_args(parser)    

add_args(None, parser)
add_miner_args(None, parser)
```

```python
#Change some default parameters in the config to move into mock
for index, action in enumerate(parser._actions):
    if action.dest in ['mock', 'neuron.should_force_model_loading']:
        parser._actions[index].default = 'True' 
    if action.dest in ['neuron.timeout', 'protein.max_steps']:
        parser._actions[index].default = 10
    if action.dest == 'protein.pdb_id':
        parser._actions[index].default = 'test_pdb'
    if action.dest == 'neuron.suppress_cmd_output':
        parser._actions[index].default = 'True'
    if action.dest == 'blacklist.allow_non_registered':
        parser._actions[index].default = 'True'

test_config = bt.config(parser)
```

```python
#create a testfoldingminer to not run anything gromacs, just some waits. 
class TestFoldingMiner(FoldingMiner):
    def configure_commands(self): 
        commands = [
            'echo "command 0"'
            'echo sleep 5'
            'echo "command 1"'
            'echo sleep 5'
            'echo "command 2"'
            'echo sleep 5'
            'echo "command 3"'
        ]

        state_commands = {
            "nvt": commands[:2],
            "npt": commands[2:4],
            "md_0_1": commands[4:],
        }

        return state_commands
```

```python
miner = TestFoldingMiner(config = test_config)
```
